MODELLING-

MODEL SELECTION-

For this dataset I would be using ANN(Artificial Neural Network) for modelling and predictions . I have chosen ANN because it is a better mode of modelling datasets and we can also get to know our accuracy in this type .

Importing the required libraries

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

We need to standardize the x_values

```python
x_values = dataframe[['Year','Child Population']]
standardise = StandardScaler() 
x_values = standardise.fit_transform(x_values)
x_values_df = pd.DataFrame(x_values)
```

Now we standardise the y_values

```python
y_values = dataframe[['Value']]
y_values = standardise.fit_transform(y_values)
```

Now training and testing data with size 0.1

```python
x_train, x_test, y_train, y_test = train_test_split(x_values,y_values,test_size=0.1,random_state=0)
print("Number of rows in x_train:", x_train.shape[0])
print("Number of rows in x_test:", x_test.shape[0])
print("Number of rows in y_train:", y_train.shape[0])
print("Number of rows in y_test:", y_test.shape[0])
```

Now bulding ANN

```python
model = Sequential()
#input layer with 6 nodes
model.add(Dense(6,input_dim=2,activation="relu"))
#hidden layers with 6 nodes each
model.add(Dense(6,activation="relu"))
model.add(Dense(6,activation="relu"))
#output layer with 1 node 
model.add(Dense(1))
model.compile(optimizer="adam",loss='mean_squared_error')
#the summary of the model
model.summary()
```

Now we can finally train the model with .fit() function

```python
model.fit(x_train,y_train,epochs=200,shuffle=True,validation_data=(x_test,y_test))
```

Now I have trained my model and for training this model , I used the 3 parameters (year , child population , value) and I only took 0.1 amount of data to test and rest 0.9 was used for training and validation data.

EVALUATION-

MODEL ACCURACY-

So as this is a regression problem so the loss at the end of the trained neural network was 0.0064 
So the accuracy 100 - 0.64 = 99.36% (0.0064 would be equal to 0.64%)

LINE GRAPH TO CHECK OVERLAPPING

```python
y_predict = model.predict(x_test)
print(y_predict)
```

```python
x = y_test
y = y_predict
plt.plot(x,y,c="red")
plt.xlabel("y_test value",fontsize=15)
plt.ylabel("y_predict value",fontsize=15)
plt.title("DIFFERENCE BETWEEN ACTUAL AND PREDICTED")
plt.show()
```
